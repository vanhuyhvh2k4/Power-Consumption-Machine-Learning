# Pre Trainning

## T√°ch c√°c ƒë·∫∑c tr∆∞ng v√† target

```python
features = ['Temperature', 'Humidity', 'WindSpeed', 'GeneralDiffuseFlows', 'Hour']
target = 'PowerConsumption_Zone1'

X = df[features]
y = df[target]

print("‚úÖ ƒê√£ t√°ch ƒë·∫∑c tr∆∞ng v√† m·ª•c ti√™u:")
print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Chia t·∫≠p d·ªØ li·ªáu train v√† test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

## Chu·∫©n ho√° d·ªØ li·ªáu

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("‚úÖ ƒê√£ chu·∫©n h√≥a d·ªØ li·ªáu:")
print("X_train_scaled shape:", X_train_scaled.shape)
print("V√≠ d·ª•:")
print(X_train_scaled[:5])
```

# Train m√¥ h√¨nh

## ƒê·ªãnh nghƒ©a 3 model v√† d√πng GridSearchCV t√¨m ra c√°c tham s·ªë t·ªët nh·∫•t cho model

```python
models = {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {
            'fit_intercept': [True, False]
        }
    },
    'RandomForest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100],             # √çt h∆°n nh∆∞ng v·∫´n ƒë·ªß ki·ªÉm tra
            'max_depth': [None, 10],               # C√≥ ho·∫∑c kh√¥ng gi·ªõi h·∫°n ƒë·ªô s√¢u
            'min_samples_split': [2, 5],           # TƒÉng nh·∫π ƒë·ªÉ ki·ªÉm so√°t overfitting
            'min_samples_leaf': [1, 2],            # ƒê∆°n gi·∫£n nh∆∞ng quan tr·ªçng
            'bootstrap': [True]                    # Gi·ªØ ƒë∆°n gi·∫£n, kh√¥ng th·ª≠ False
        }
    },
    'XGBoost': {
        'model': XGBRegressor(random_state=42, verbosity=0),
        'params': {
            'n_estimators': [100, 200],            # Hai m·ª©c ƒë·ªß r√µ r√†ng
            'learning_rate': [0.05, 0.1],          # H·ªçc ch·∫≠m h∆°n m·ªôt ch√∫t
            'max_depth': [3, 6],                   # ƒê·ªô ph·ª©c t·∫°p trung b√¨nh
            'subsample': [0.8, 1.0],               # Gi·∫£m overfitting
            'colsample_bytree': [0.8, 1.0],        # C√¢n b·∫±ng ƒë·ªô ƒëa d·∫°ng c√¢y
            'gamma': [0, 0.1],                     # Ph·∫°t ph√¢n t√°ch y·∫øu
            'reg_alpha': [0, 0.1],                 # L1 regularization
            'reg_lambda': [1, 1.5]                 # L2 regularization
        }
    }
}
```

## Trainning model

```python
best_models = {}

for name, config in models.items():
    print(f"\nTuning {name}...")

    start = time.time()

    # T√≠nh s·ªë l∆∞·ª£ng t·ªï h·ª£p tham s·ªë
    total_combinations = len(list(ParameterGrid(config['params'])))

    # D√πng GridSearchCV n·∫øu m√¥ h√¨nh ƒë∆°n gi·∫£n ho·∫∑c s·ªë t·ªï h·ª£p nh·ªè
    if name in ['LinearRegression', 'RandomForest'] or total_combinations <= 20:
        print(f"Using GridSearchCV ({total_combinations} combinations).")
        search = GridSearchCV(
            estimator=config['model'],
            param_grid=config['params'],
            scoring='neg_mean_squared_error',
            cv=3,
            verbose=0,
            n_jobs=-1
        )
    else:
        print(f"Using RandomizedSearchCV (random 20 out of {total_combinations}).")
        search = RandomizedSearchCV(
            estimator=config['model'],
            param_distributions=config['params'],
            n_iter=20,
            scoring='neg_mean_squared_error',
            cv=3,
            verbose=0,
            random_state=42,
            n_jobs=-1
        )

    # Hu·∫•n luy·ªán
    search.fit(X_train_scaled, y_train)
    end = time.time()

    best_models[name] = {
        'best_estimator': search.best_estimator_,
        'best_params': search.best_params_,
        'best_score': -search.best_score_,
        'train_time_sec': round(end - start, 2)
    }

    print(f"{name} done in {best_models[name]['train_time_sec']} seconds.")

# In k·∫øt qu·∫£
print("\n==== Best Models Summary ====")
for name, result in best_models.items():
    print(f"\n{name}")
    print(f"Best Params: {result['best_params']}")
    print(f"Best Score (MSE): {result['best_score']:.4f}")
    print(f"Training Time: {result['train_time_sec']}s")

# Print best models and their scores
for name, result in best_models.items():
    print(f"{name}:")
    print(f"Best Params: {result['best_params']}")
    print(f"Best Score (MSE): {result['best_score']}")
```

# So s√°nh model v·ªõi d·ªØ li·ªáu th·ª±c

```python
# L·∫•y m√¥ h√¨nh t·ªët nh·∫•t t·ª´ best_models
best_lr = best_models['LinearRegression']['best_estimator']
best_rf = best_models['RandomForest']['best_estimator']
best_xgb = best_models['XGBoost']['best_estimator']

# D·ª± ƒëo√°n
y_pred_lr = best_lr.predict(X_test_scaled)
y_pred_rf = best_rf.predict(X_test_scaled)
y_pred_xgb = best_xgb.predict(X_test_scaled)

# T√≠nh c√°c ch·ªâ s·ªë
def evaluate(y_true, y_pred):
    return {
        'R2': r2_score(y_true, y_pred),
        'MAE': mean_absolute_error(y_true, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred))
    }

metrics_lr = evaluate(y_test, y_pred_lr)
metrics_rf = evaluate(y_test, y_pred_rf)
metrics_xgb = evaluate(y_test, y_pred_xgb)

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(12, 5))
plt.plot(y_test.values[:100], label='Th·ª±c t·∫ø', linewidth=2)
plt.plot(y_pred_lr[:100], label=f'Linear Reg (R¬≤={metrics_lr["R2"]:.2f})', linestyle='--')
plt.plot(y_pred_rf[:100], label=f'Random Forest (R¬≤={metrics_rf["R2"]:.2f})', linestyle='--')
plt.plot(y_pred_xgb[:100], label=f'XGBoost (R¬≤={metrics_xgb["R2"]:.2f})', linestyle='--')
plt.title('So s√°nh d·ª± ƒëo√°n v·ªõi d·ªØ li·ªáu th·ª±c t·∫ø (100 ƒëi·ªÉm ƒë·∫ßu)')
plt.xlabel('Th·ªùi ƒëi·ªÉm')
plt.ylabel('Ti√™u th·ª• ƒëi·ªán nƒÉng')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# In th√™m b·∫£ng ƒë√°nh gi√°
print("\n=== ƒê√°nh gi√° ƒë·ªô ch√≠nh x√°c tr√™n t·∫≠p test ===")
for name, metrics in zip(['LinearRegression', 'RandomForest', 'XGBoost'],
                         [metrics_lr, metrics_rf, metrics_xgb]):
    print(f"\n{name}")
    print(f"R¬≤:   {metrics['R2']:.4f}")
    print(f"MAE:  {metrics['MAE']:.4f}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
```
